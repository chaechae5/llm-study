# Perceptron

```python
class Perceptron:
    def __init__(self, w1, w2, theta):
        self.w1 = w1
        self.w2 = w2
        self.theta = theta

    def run(self, x1, x2):
        tmp = x1*self.w1 + x2*self.w2
        if tmp <= self.theta:
            return 0
        else:
            return 1
```

## Logic gate

```python
def test(gate):
    inputs = [ [0, 0], [1, 0], [0, 1], [1, 1] ]
    for input in inputs:
        x1 = input[0]
        x2 = input[1]

        print("{} {} {}".format(x1, x2, gate.run(x1, x2)))
```

### AND gate

| *x1* | *x2* | *y* |
| --- | --- | --- |
| 0 | 0 | 0 |
| 1 | 0 | 0 |
| 0 | 1 | 0 |
| 1 | 1 | 1 |



```python
and_gate = Perceptron(0.5, 0.5, 0.7)
```

```python
test(and_gate)
```
